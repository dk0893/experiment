```python
%matplotlib notebook
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import animation, rc
```

```python
def set_figure( ax, xs, ys ):
    #ax.clear()
    ax.tick_params(labelbottom=False, labelleft=False, labelright=False, labeltop=False)
    ax.set_xticks(range(xs))
    ax.set_yticks(range(ys))
    ax.set_xlim(0, xs)
    ax.set_ylim(0, ys)
    ax.grid(True)
```

```python
color_list = ['red', 'white', 'green']
cmap = matplotlib.colors.LinearSegmentedColormap.from_list( 'colormap_name', color_list )

fig = plt.figure( figsize=(6, 4) )
ax = fig.add_subplot( 111 )
#fig, ax = plt.subplots() # こっちでもいい

artists = []
np_seq = np.arange(6).reshape(2, 3)
lst_value = [np_seq, np_seq / 6, np.flipud(np_seq)]
lst_vmin  = [None, None, None]
lst_vmax  = [None, None, None]
lst_text  = ["0: np_seq", "1: np_seq / 6", "2: np.flipud(np_seq)"]

for val, txt, vmin, vmax in zip(lst_value, lst_text, lst_vmin, lst_vmax):
  #ax = fig.add_subplot(111)
  set_figure( ax, 3, 2 )

  print( val )

  art_mesh = ax.pcolormesh( val, cmap=cmap, vmin=vmin, vmax=vmax )
  art_text = ax.text( 0, 2.1, txt )
  artists.append( [art_mesh, art_text] )
```

```python
print( artists )
```

```python
anim = animation.ArtistAnimation( fig, artists, interval=2000, repeat_delay=1000 )
anim.save( "anim.gif" )

# Google Colaboratoryの場合必要
rc('animation', html='jshtml')
#plt.close()
anim
```
